```python
from urllib.request import urlopen
from bs4 import BeautifulSoup

import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None)
```

```python
NUMOFROWS = "10"       ## 정보 출력 개수
OBSID = "6000"      ## 산 정보
TM = "202001010000"           ## 데이터 수집 시간정보 기입 (YYYYMMDDHHmm)
LOCAL_AREA = "13"  ## 전라남도
SERVICE_KEY = "4ypIStp6BalVUJZ4M%2F9weoBy6F9ft9Sq0AIorhAJvXb9vsnHw%2BwQmvF6xrmz76vulRCPHYykuf8Ypf9oqWLR3A%3D%3D"

url = f'http://apis.data.go.kr/1400377/mtweather/mountListSearch'\
f'?serviceKey={SERVICE_KEY}'\
f'&numOfRows={NUMOFROWS}&pageNo=1&localArea={LOCAL_AREA}'\
f'&obsid=&tm={TM}&_type=xml'
```

```python
result = urlopen(url)
house = BeautifulSoup(result, 'lxml-xml')
te = house.find_all('item')

te
```

```python
datas = []

data_col = ["무게식_강우량", "10m습도", "2m습도", "지역코드", "지점번호", "산이름", "기압", "전도식_강우량", "관측시간",
            "10m기온", "2m기온", "지면온도", "10m풍향", "10m풍향방위", "2m풍향", "2m풍향방위", "10m풍속",
           "2m풍속"]

for i in range(len(te)):
    cprn = te[i].cprn.string.strip()
    hm10m = te[i].hm10m.string.strip()
    hm2m = te[i].hm2m.string.strip()
    localarea = te[i].localarea.string.strip()
    obsid = te[i].obsid.string.strip()
    obsname = te[i].obsname.string.strip()
    pa = te[i].pa.string.strip()
    rn = te[i].rn.string.strip()
    tm = te[i].tm.string.strip()
    tm10m = te[i].tm10m.string.strip()
    tm2m = te[i].tm2m.string.strip()
    ts = te[i].ts.string.strip()
    wd10m = te[i].wd10m.string.strip()
    wd10mstr = te[i].wd10mstr.string.strip()
    wd2m = te[i].wd2m.string.strip()
    wd2mstr = te[i].wd2mstr.string.strip()
    ws10m = te[i].ws10m.string.strip()
    ws2m = te[i].ws2m.string.strip()
    
    data = [cprn, hm10m, hm2m, localarea, obsid, obsname, pa, rn, tm,
            tm10m, tm2m, ts, wd10m, wd10mstr, wd2m, wd2mstr, ws10m,
           ws2m]
    
    datas.append(data)
```

```python
datas
```

```python
df = pd.DataFrame(datas, columns = data_col)
df.head()
```

```python
df.to_csv("산악기상정보.csv", encoding = 'euc-kr', index = False)
```
