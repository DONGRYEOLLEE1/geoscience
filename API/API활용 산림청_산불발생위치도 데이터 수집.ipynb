```python
from urllib.request import urlopen
from bs4 import BeautifulSoup

import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None)
pd.set_option('display.max_info_columns', 20)
```

```python
SEARCHSTDT = '20110101'
SEARCHEDDT = '20110201'
SERVICE_KEY = '서비스키'

url = f'http://openapi.forest.go.kr/openapi/service/forestDisasterService/frstFireOpenAPI?'\
f'searchStDt={SEARCHSTDT}&searchEdDt={SEARCHEDDT}&numOfRows=50&ServiceKey={SERVICE_KEY}'
```

```python
result = urlopen(url)
house = BeautifulSoup(result, 'lxml-xml')
te = house.find_all('item')

te
```

```python
datas = []
xml_col = ['diravg', 'dirmax', 'dmgarea', 'dmgmoney', 'exintgtm', 'extingdt', 'humidcurr', 'humidmin', 'humidrel', 'ocurcause', 
          'ocurdo', 'ocurdt', 'ocuremd', 'ocurgm', 'ocurjibun', 'ocurri', 'ocursgg', 'ocuryoil', 'ownersec', 'rainamount', 'raindays',
          'riskavg', 'riskmax', 'tempavg', 'windavg', 'windmax']
data_col = ["평균풍향", "최대풍향", "피해면적", "피해액", "진화소요시간(분)", "진화일시(년월일시분)", "실효습도", "최소습도", "상대습도",
           "발생세부원인", "발생장소(시도)", "발생일시", "발생장소(읍면동)", "발생장소(관서)", "발생장소(지번)", "발생장소(리)",
           "발생장소(시군구)", "발생일시(요일)", "소유구분", "최종강우량", "강우경과일수", "산불위험지수(평균)", "산불위험지수(최대)",
           "평균기온", "평균풍속", "최대풍속"]

for i in range(len(te)):
    평균풍향 = te[i].diravg.string.strip()
    최대풍향 = te[i].dirmax.string.strip()
    피해면적 = te[i].dmgarea.string.strip()
    피해액 = te[i].dmgmoney.string.strip()
    진화소요시간_분 = te[i].exintgtm.string.strip()
    진화일시_년월일시분 = te[i].extingdt.string.strip()
    실효습도 = te[i].humidcurr.string.strip()
    최소습도 = te[i].humidmin.string.strip()
    상대습도 = te[i].humidrel.string.strip()
    발생세부원인 = te[i].ocurcause.string.strip()
    발생장소_시도 = te[i].ocurdo.string.strip()
    발생일시 = te[i].ocurdt.string.strip()
    발생장소_읍면동 = te[i].ocuremd.string.strip()
    발생장소_관서 = te[i].ocurgm.string.strip()
    발생장소_지번 = te[i].ocurjibun.string.strip()
    발생장소_리 = te[i].ocurri.string.strip()
    발생장소_시군구 = te[i].ocursgg.string.strip()
    발생일시_요일 = te[i].ocuryoil.string.strip()
    소유구분 = te[i].ownersec.string.strip()
    최종강우량 = te[i].rainamount.string.strip()
    강우경과일수 = te[i].raindays.string.strip()
    산불위험지수_평균 = te[i].riskavg.string.strip()
    산불위험지수_최대 = te[i].riskmax.string.strip()
    평균기온 = te[i].tempavg.string.strip()
    평균풍속 = te[i].windavg.string.strip()
    최대풍속 = te[i].windmax.string.strip()
    
    data = [평균풍향, 최대풍향, 피해면적, 피해액, 진화소요시간_분, 진화일시_년월일시분, 실효습도, 최소습도, 상대습도,
           발생세부원인, 발생장소_시도, 발생일시, 발생장소_읍면동, 발생장소_관서, 발생장소_지번, 발생장소_리,
           발생장소_시군구, 발생일시_요일, 소유구분, 최종강우량, 강우경과일수, 산불위험지수_평균, 산불위험지수_최대,
           평균기온, 평균풍속, 최대풍속]
    
    datas.append(data)
```

```python
df = pd.DataFrame(datas, columns = data_col)
df.head(3)
```

```python
# 전처리
df['피해면적'] = df['피해면적'].astype(float)
df['피해액'] = df['피해액'].astype(int)
df['진화소요시간(분)'] = df['진화소요시간(분)'].astype(int)
df['실효습도'] = df['실효습도'].astype(float)
df['최소습도'] = df['최소습도'].astype(float)
df['상대습도'] = df['상대습도'].astype(float)
df['최종강우량'] = df['최종강우량'].astype(float)
df['강우경과일수'] = df['강우경과일수'].astype(int)
df['산불위험지수(평균)'] = df['산불위험지수(평균)'].astype(float)
df['산불위험지수(최대)'] = df['산불위험지수(최대)'].astype(float)
df['평균기온'] = df['평균기온'].astype(float)
df['평균풍속'] = df['평균풍속'].astype(float)
df['최대풍속'] = df['최대풍속'].astype(float)
```
